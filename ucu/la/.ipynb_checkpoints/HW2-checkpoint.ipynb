## Problem 1


Determine whether the given planes are:

parallel:

1) $4x-y+2z=5$ and $7x-3y+4z=8$;


2) $x-4y-3z-2=0$ and $3x-12y-9z-7=0$.

perpendicular:

1) $3x-y+z=0$ and $x+2z=-1$;

2) $x-2y+3z=4$ and $-2x+5y+4z=-1$.


### Solution 
Let's take normal vecotrs of the planes. The two vectors are not proportional to each other, thus the planes are not parallel and they intersect each other.
If the dot product of the two normals is 0, then the two planes are perpendicular

#### Parallel:

1) $4x-y+2z=5$ and $7x-3y+4z=8$;

$\frac{4}{7} \neq \frac{-1}{-3} \neq \frac{2}{4}$ -> **Planes are not parallel**

2) $x-4y-3z-2=0$ and $3x-12y-9z-7=0$.

$\frac{1}{3} = \frac{-4}{-12} = \frac{-3}{-9}$-> **Planes are parallel**

#### Perpendicular:

1) $3x-y+z=0$ and $x+2z=-1$;

$\left(\begin{matrix}
3 & -1 & 1
\end{matrix}\right) *
\left(\begin{matrix}
1 \\
0 \\
2
\end{matrix}\right) = 
\left(\begin{matrix}
5
\end{matrix}\right)
$
-> **Planes are not perpendicular**

2) $x-2y+3z=4$ and $-2x+5y+4z=-1$.

$
\left(\begin{matrix}
1 & -2 & 3
\end{matrix}\right)
*
\left(\begin{matrix}
-2 \\
5 \\
4
\end{matrix}\right)
=
\left(\begin{matrix}
0
\end{matrix}\right)
$
-> **Planes are perpendicular**


## Problem 2

a) Let $W$ be the plane in $R^3$ given by the equation $x-2y-3z=0$. Find parametric equations for $W^\perp$.


b) Let $W$ be the line in $R^3$ with parametric equations $x=2t$, $y=-5t$, $z=4t$. Find an equation for $W^\perp$.


c) Let $W$ be the intersection of the two planes $x+y+z=0$ and $x-y+z=0$ in $R^3$. Find an equation for $W^\perp$.

### Solution

#### Item a
$x-2y-3z=0$

$dim(W)=2$, so $dim(W^\perp)=1$

$M = {(x, y, z)^T ∈ R^3 : x-2y-3z=0}$

$x-2y-3z=0 \rightarrow x=2y+3z$

$(x, y, z)^T = (2y + 3z, y, z)^T = y(2, 1, 0)^T + z(3, 0, 1)^T $

$ A = \left( \begin{matrix}
2 & 3 \\
1 & 0 \\
0 & 1
\end{matrix}\right)
$

$C(A)$ is the orthogonal complement of it's left nullspace $N(A^T)$ 


$ A^T = \left(\begin{matrix}
2 & 1 & 0 \\
3 & 0 & 1
\end{matrix}\right)
$

$N(A^T)$ is solution of $A^Tx=0$, 

Solving $A^Tx=0$ with gauss elimination we get a solution set $x_3 \left(\begin{matrix}
\frac{-1}{3} \\
\frac{2}{3} \\
1
\end{matrix}\right)$

**As parametric:** $x=\frac{-1}{3}t$, $y=\frac{2}{3}t$, $z=t$

Note: We can see that our solution is a normal vector of the plane divided by -3 

#### Item b
As we can see from the 1st task - coeficients of the parametric equation is normal vector to the plane. So we could construct a plane equation from the given parametric equation: 

**Equation:** $2x-5y+4z=0$

#### Item c
W(line) is intersection of the two planes. Let's find vector of the direction of the intersection line by taking cross product of normal vectors of the planes. 

$a=(1, 1, 1)^T$

$b=(1,-1,1)^T$

$a x b = \left|\begin{matrix}
i & j & k \\
1 & 1 & 1 \\
1 & -1 & 1
\end{matrix}\right| = 2i -2k$

$c = (2, 0, -2)$

We are looking for $W^\perp$. As c is direction vector of W -> c is a normal to the $W^\perp$. So we could construct a plane equation from the biven parametric equation

**Equation:** $2x-2z=0$

### Problem 3
a) Find the distance from the point $P=(1,1,0)$ to the line $\frac{x-1}{2}=\frac{y-2}{1} = \frac{z+1}{2}$. 

b) Let $\pi$ be a plane given by the equation $ax + by + cz +d = 0$ and $P(x_0,y_0,z_0)$ be a point outside it. Prove that the distance from $P$ to $\pi$ is given by the formula
		$
			\frac{|ax_0 + by_0 + cz_0 + d|}{\sqrt{a^2 + b^2 + c^2}}.
		$
		
Hint: if $Q$ is the point on $\pi$ realizing the distance, then $\overrightarrow{PQ}$ is collinear to $\mathbf{n}=(a,b,c)$ (why?). Take now any point $Q'$ on $\pi$ and find a projection of $\overrightarrow{PQ'}$ onto direction 
$\mathbf{n}$

c) Find the distance between the point~$P=(1,0,1)$ and the plane~$2x+2y-z=2$.


### Solution

#### Item a

From line equation find:

$s=(2,1,2)^T$ - direction vector of the line
$M=(1, 2, -1)$ - point on the line

$PM = (0, 2, -2)$

$PM x s = \left|\begin{matrix}
i & j & k \\
0 & 2 & -2 \\
2 & 1 & 2
\end{matrix}\right| = -4*j-4*k+6i $

$d = \frac{|PM x s|}{|s|} = \frac{68}{9} = 7\frac{5}{9}$

**Distance:** $7\frac{5}{9}$


#### Item b
Let $N = (a, b, c)$ is normal vector to the plane. The shortest distance from a point to a plane is along a line perpendicular to the plane. So, the distance from point P to the plane is along a line parallel to the normal vector. 
Let's take $P=(x_1, y_1, z_1)$ and $Q=(x_0, y_0, z_0)$ -> $QP=(x_1-x_0, y_1-y_0, z_1-z_0)$

We want to get a length of a projection of QP onto normal unit vector;

$n = \frac{N}{||N||}$

Also, we have initial $Ax + By + Cz + D =0$ -> $D = -Ax -By -Cz$

Puting things together and calculating length of a projeciton:

$D = |QP * n| = \frac{|A(x_1-x_0) + B(y_1-y_0) + C(z_1-z_0)|}{\sqrt{A^2 + B^2 + C^2}} = \frac{|Ax_1 + By_1 + Cz_1 + D|}{\sqrt{A^2 + B^2 + C^2}}$

$-Ax -By -Cz$ is replaced by $D$ after opening the brackets. This is possible only because we know that Q is point on the plane, so the distance from P to plane doesn't depend on the any point.


#### Item c
Using formula from b

$2x+2y-z=2 \rightarrow 2x+2y-z-2=0$

$D = \frac{|2 * 1 + 2 * 0 -1 * 1 - 2|}{\sqrt{2^2 + 2^2 +(-1)^2}} = \frac{|-1|}{\sqrt{9}}=\frac{1}{3}$


**Answer:** $D=\frac{1}{3}$

### Problem 9
Assume $u_1$ and $u_2$ are two orthogonal vectors in $R^n$ and set $a_1 = u_1$, $a_2 = u_1 + \varepsilon u_2$ for $\varepsilon>0$. Let also $A$ be the matrix with columns $a_1$ and $a_2$ and $b$ a vector linearly independenet of $a_1$ and $a_2$. In this problem, we discuss the least square solution to the system $Ax = b$ as $\varepsilon\to0$. 

a) Calculate the matrix $A^\top A$, its inverse, and then $\hat{x} = (A^T A)^{-1}A^Tb$ explicitly. Show that $\hat{x}$ explodes as $\varepsilon\to0$.

b) Calculate the projection $A\hat{x}$ of $b$ onto $\operatorname{col}(A)$ and check that it does not depend on~$\varepsilon>0$. Explain the result.



### Solution

$A=\left(\begin{matrix}
u & u+a*v
\end{matrix}\right)$


$A^TA = \left(\begin{matrix}
u^2 & u^2+a*u*v \\
u^2+a*u*v & u^2+a^2*v^2+2*a*u*v
\end{matrix}\right)$

As v and u orthogonal vectors matrix can be simplified.

$A^TA = \left(\begin{matrix}
u^2 & u^2 \\
u^2 & u^2+a^2*v^2
\end{matrix}\right)$

$(A^TA)^{-1} = \left(\begin{matrix}
\frac{u^2+a^2*v^2}{a^2*u^2*v^2} & \frac{-1}{a^2*v^2} \\
\frac{-1}{a^2*v^2} & \frac{1}{a^2*v^2}
\end{matrix}\right)$


$\hat{x} = (A^TA)^{-1}A^Tb = \left(\begin{matrix}
\frac{u^2+a^2*v^2}{a^2*u^2*v^2} & \frac{-1}{a^2*v^2} \\
\frac{-1}{a^2*v^2} & \frac{1}{a^2*v^2}
\end{matrix}\right) \left(\begin{matrix}
u \\
u+a*v
\end{matrix}\right)\left(\begin{matrix}
b
\end{matrix}\right) = \left(\begin{matrix}
\frac{-u+a*v}{a*u*v} \\
\frac{1}{a*v}
\end{matrix}\right)\left(\begin{matrix}
b
\end{matrix}\right)=\left(\begin{matrix}
\frac{-b*u+a*b*v}{a*u*v} \\
\frac{b}{a*v}
\end{matrix}\right)$


As we can see - $a$ is present in both denominators. If $a \rightarrow 0$ fraction will go to the infinity and explode (fraction $ \rightarrow \inf$)



#### b 

$\left(\begin{matrix}
u & u+a*v
\end{matrix}\right) \left(\begin{matrix}
\frac{-b*u+a*b*v}{a*u*v} \\
\frac{b}{a*v}
\end{matrix}\right) = \left(\begin{matrix}
2*b
\end{matrix}\right)$

### Problem 12
a) Find the unit vector $u\in R^2$ such that the ouseholder reflection $Q_{u}:=I - 2uu^T$ maps the vector~$(1,2)^T$ onto a vector collinear to $(1,0)^T$

b) explain how $Q_{u}$ helps to derive the $QR$ factorization of the matrix (a) of Problem 11.

c) Find the $QR$-factorization of matrices in (b) and (c) of Problem 11 using the Householder reflections approach.
	


### Solution
#### Item  a

We are given a vector x and we want to find a reflection
that transforms x into a direction parallel to some unit vector y. 

$v = x - ||x||y$

$u = \frac{v}{||v||}$

$Q = I - 2uu^T = I - 2\frac{vv^T}{v^Tv} $

$v = (1-\sqrt{5}, 2)^T$

$u = (\frac{1-\sqrt{5}}{\sqrt{10-2\sqrt{5}}}, \frac{2}{\sqrt{10-2\sqrt{5}}})^T$

$Q = I - 2\frac{\left(\begin{matrix}
-2*\sqrt{5}+6 & -2*\sqrt{5}+2 \\
-2*\sqrt{5}+2 & 4
\end{matrix}\right)}{\left(\begin{matrix}
-2*\sqrt{5}+10
\end{matrix}\right)} =  I - \left(\begin{matrix}
\frac{-\sqrt{5}+5}{5} & \frac{-2*\sqrt{5}}{5} \\
\frac{-2*\sqrt{5}}{5} & \frac{\sqrt{5}+5}{5}
\end{matrix}\right) = \left(\begin{matrix}
\frac{\sqrt{5}}{5} & \frac{2*\sqrt{5}}{5} \\
\frac{2*\sqrt{5}}{5} & \frac{-\sqrt{5}}{5}
\end{matrix}\right)$

Check the result: 
$ Qx = \left(\begin{matrix}
\frac{5^\left(\frac{1}{2}\right)}{5} & \frac{2*5^\left(\frac{1}{2}\right)}{5} \\
\frac{2*5^\left(\frac{1}{2}\right)}{5} & \frac{-5^\left(\frac{1}{2}\right)}{5}
\end{matrix}\right)x \left(\begin{matrix}
1 \\
2
\end{matrix}\right) = \left(\begin{matrix}
\sqrt{5} \\
0
\end{matrix}\right)$ colinear to $y$

**Answer:** $u = (\frac{1-\sqrt{5}}{\sqrt{10-2\sqrt{5}}}, \frac{2}{\sqrt{10-2\sqrt{5}}})^T$

#### Item b
A Householder reflection takes a vector and reflects it about some plane or hyperplane. We could use Q to reflect a vector in a way that all coordinates but one will disappear. Applying reflections starting tfrom the 1st column of matrix A, reflecting the 1st vector and taking minor of A_11 we can reduce matrix to upper triangular form(R). When reducing the A matrix we still want to multiply all Q's so we have to extend reduced matrix with 1 on diagonals($R=Q_n...Q_1A$). To get Q we have to multiply matrices intermediate transposed Qs. ($Q = Q_1^T...Q_n^T$)

This method has greater numerical stability than the Gram–Schmidt method



#### Item c

To apply the householder reflection to get QR we will use the following algorithm:

1) Find the reflection that transforms the first column of the matrix into $||a_1|| e_1$

2) Save reflection matrix, Multiply A from the left by reflection matrix

3) Reduce A (remove 1 row and 1 column) and repeat for the reduced matrix until we get last reflection matrix

4) Calculate $R$ and $Q$ as mentioned in **b**
