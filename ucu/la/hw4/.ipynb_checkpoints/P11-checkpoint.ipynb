```python
import numpy as np
from matplotlib import pyplot as plt
```

#### A) Generating the data: Build covariance matrix from $\sigma^2$ and $\rho$. Data is generated with numpy multivariate normal. 
#### B) Covariance matrix is constructed manually (covm variable)

```python
def generate_data(p):
    np.random.seed(42)
    means = [1, 2]
    variances = [4, 9]
    covm = np.array([[4, 2 * 3 * p], [2 * 3 * p, 9]])
    data = np.random.multivariate_normal(means, covm, 100)
    return data
```

#### B) eigenvalues and eigenvectors are found using numpy.eig function
#### C) Fraction of variance is calculated using eigenvalues. Fraction calculated as eigvalue_i / sum(eigvalues)

```python
def generate_and_pca(p):
    data = generate_data(p)
    z_mean = data - np.mean(data, axis=0)
    ecov = np.matmul(z_mean.T, z_mean)
    evalues, evectors = np.linalg.eig(ecov)
    return evalues, evectors
```

```python
p_range = [x / 100 for x in range(-1000, 1000, 1)]
```

```python
var_expl = np.vstack([(evalues / np.sum(evalues) * 100) 
                      for evalues, evectors in [generate_and_pca(x) 
                                                for x in p_range]])
```

#### We can see that for p=-1 and p=1 our variable x and y  have follow interesting pattern and can be explained very well by 1 principal component 

```python
for i in range(-4, 4):
    data = generate_data(i)
    plt.scatter(x=data[:, 0], y = data[:, 1])
    plt.title("P={}".format(i))
    plt.show()
```

#### C, D
Here we have plot of % for each principal component. On X axis we have $\rho$ value (-10, 10) and on Y axis we have % of significance. As we have two components, they sum up to 100%. 

We can see that for P=-1 and P=1 and area that close to that values, one component can explain the data pretty good. Bigger distance to range (-1, 1) results in almost equal component significance (between 40% and 60%)

```python
width=1
plt.figure(figsize=(10, 5))
plt.bar(x=p_range, height=var_expl[:, 0], color='blue', label='First component', width=width)
plt.bar(x=p_range, height=var_expl[:, 1], bottom=var_expl[:, 0], color='orange', label='Second component', width=width)
plt.legend()
plt.show()
```
