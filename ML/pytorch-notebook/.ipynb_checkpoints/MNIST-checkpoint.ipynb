```python
import torch
import os
import numpy as np
from torch.autograd import Variable
from skimage import io, transform
import torch.nn as nn
import torch.nn.functional as F
from sklearn import preprocessing
import pandas as pd
import matplotlib.pyplot as plt
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, utils
from skimage.transform import resize
```

```python
train_set = pd.read_csv('mnist_train.csv', header=None)
```

```python
res = train_set.as_matrix()
```

```python
res
```

```python
labelToImage = list(map(lambda x: [x[0], np.array(x[1:]).reshape(28,28)], res))
```

```python
df = pd.DataFrame(labelToImage)
```

```python
df.columns =['label', 'image']
```

```python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()

        self.conv1 = nn.Conv2d(1, 5, 3, padding=1)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(5, 15, 3, padding=1)
        self.fc1 = nn.Linear(735, 28 * 28)
        self.fc2 = nn.Linear(28 * 28, 28 * 2 * 5)
#         Classes count
        self.fc3 = nn.Linear(28 * 2 * 5, 10)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.view(-1, 735)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x
```

```python
net = Net()
```

```python
loss_fn = nn.CrossEntropyLoss()
```

```python
optimizer = torch.optim.SGD(net.parameters(), lr=0.001, momentum=0.9)
```

```python
label, image = df.iloc[0]
```

```python
dims = [1, 1, 28, 28]
```

```python

for epoch in range(0, 3):
    for i in range(0, df.shape[0]):
        label, image = df.iloc[i]
        label = Variable(torch.from_numpy(np.array([int(label)])))
        image = Variable(torch.from_numpy(image).view(dims).float())
        optimizer.zero_grad()

        prediction = net(image)
        loss = loss_fn(prediction, label)
        if i % 20 == 19:
            print("Epoch: " + str(epoch) + "Iter: "+ str(i) , loss.data[0] * 100)
        loss.backward()
        optimizer.step()
    
    
```

```python
torch.save(net.state_dict(),"model-mnist2")
```

```python
test_set = pd.read_csv('mnist_test.csv', header=None)
```

```python
test = test_set.as_matrix()
```

```python
labelToImage_test = list(map(lambda x: [x[0], np.array(x[1:]).reshape(28,28)], test))
```

```python
test_df = pd.DataFrame(labelToImage_test)
```

```python
test_df.columns =['label', 'image']
```

```python
positive = 0
total = test_df.shape[0]
for i in range(0, total):
    label, image = test_df.iloc[i]
    image = Variable(torch.from_numpy(image).view(dims).float())
    prediction = torch.max(net(image), 1)[1].data[0][0]
    if prediction == label:
        positive = positive + 1
    
    
  
```
